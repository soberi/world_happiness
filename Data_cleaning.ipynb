## Importing data and adding year timestamp

#### The original data were in separate csv files with no timestamp. In order to more easily use the data in Tableau, I have added a year column to each seperate data set.

```python
import pandas as pd
```

```python
world_2015 = pd.read_csv("world_happiness_data/world_happiness_2015.csv")
year = 2015
world_2015["Year"] = year
```

```python
world_2016 = pd.read_csv("world_happiness_data/world_happiness_2016.csv")
year = 2016
world_2016["Year"] = year
```

```python
world_2017 = pd.read_csv("world_happiness_data/world_happiness_2017.csv")
year = 2017
world_2017["Year"] = year
```

```python
world_2018 = pd.read_csv("world_happiness_data/world_happiness_2018.csv")
year = 2018
world_2018["Year"] = year
```

```python
world_2019 = pd.read_csv("world_happiness_data/world_happiness_2019.csv")
year = 2019
world_2019["Year"] = year
```

## Cleaning the columns

#### Some columns were renamed in later years, so I renamed them.

```python
cols15 = ["country", "region", "happiness_rank", "happiness_score",
          "standard_error", "GDP_per_capita", "social_support",
          "health_life_expectancy", "freedom", "trust_gov_corruption",
          "generosity", "dystopia_residual", "year"]

world_2015.columns = cols15

world_2015.head()
```

```python
cols16 = ["country", "region", "happiness_rank", "happiness_score",
          "lower_confidence_int", "upper_confidence_int", 
          "GDP_per_capita", "social_support", "health_life_expectancy", 
          "freedom", "trust_gov_corruption", "generosity",
          "dystopia_residual", "year"]

world_2016.columns = cols16

world_2016.head()
```

```python
cols17 = ["country", "happiness_rank", "happiness_score", 
          "upper_confidence_int", "lower_confidence_int", 
          "GDP_per_capita", "social_support", "health_life_expectancy", 
          "freedom", "generosity", "trust_gov_corruption", 
          "dystopia_residual", "year"]

world_2017.columns = cols17

world_2017.head()
```

```python
cols18 = ["happiness_rank", "country", "happiness_score", 
          "GDP_per_capita", "social_support", "health_life_expectancy", 
          "freedom", "generosity", "trust_gov_corruption", "year"]

world_2018.columns = cols18

world_2018.head()
```

```python
cols19 = ["happiness_rank", "country", "happiness_score", 
          "GDP_per_capita", "social_support", "health_life_expectancy", 
          "freedom", "generosity", "trust_gov_corruption", "year"]

world_2019.columns = cols19

world_2019.head()
```

## Merging all years

#### After cleaning up the columns I tried to merge the data sets within Tableau, but I could not just add rows without causing some for of error. Here I have just appended the data sets using pandas.

```python
whole_world_happiness_data = world_2015.append(world_2016).append(world_2017).append(world_2018).append(world_2019).reset_index()

whole_world_happiness_data
```
